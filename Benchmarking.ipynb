```python
categories = ['roleplay', 'reasoning', 'math', 'coding', 'extraction', 'humanities', 'writing']
models = ['Chat GPT-4', 'Gemma', 'claude', 'Llama-3-70B', 'Mistral']
scores = {
    'roleplay': [8, 7, 9, 7, 8],
    'reasoning': [10, 10, 2, 2, 2],
    'math': [2, 1, 1, 1, 2],
    'coding': [8, 6, 8, 6, 7],
    'extraction': [10, 9, 8, 10, 7],
    'humanities': [10, 7, 9, 6, 7],
    'writing': [10, 10, 10, 9, 10]
}

average_scores = {}

for model in models:
    total_score = 0
    for category in categories:
        total_score += scores[category][models.index(model)]
    average_scores[model] = total_score / len(categories)
best_model = max(average_scores, key=average_scores.get)
print("Average Scores:")
for model, avg_score in average_scores.items():
    print(f"{model}: {avg_score:.2f}")

print(f"\nOverall Best Model: {best_model}")

colors = ['skyblue', 'orange', 'green', 'red', 'purple']
plt.figure(figsize=(10, 6))
bars = plt.bar(average_scores.keys(), average_scores.values(), color=colors)
plt.title('Average Scores of Models')
plt.xlabel('Models')
plt.ylabel('Average Scores')
plt.ylim(0, 10) 
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', va='bottom') 

plt.show()
```
